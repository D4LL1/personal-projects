```python
import yfinance as yf
import matplotlib.pyplot as plt

# Zdefiniowanie daty początkowej i końcowej
start_date = '2009-01-01'
end_date = '2024-11-30'

# Pobranie danych dla Bitcoina
data = yf.download('BTC-USD', start=start_date, end=end_date)['Adj Close']

# Tworzenie wykresu
plt.figure(figsize=(10, 6))

# Rysowanie wykresu dla Bitcoina
plt.plot(data, label='Bitcoin', color='orange')  # Zmieniono nazwę na Bitcoin i kolor na pomarańczowy

# Dodanie legendy
plt.legend()

# Dodanie tytułu i etykiet osi
plt.title('Cena Bitcoina (BTC) w latach 2009-2024')
plt.xlabel('Data')
plt.ylabel('Cena zamknięcia (USD)')

# Wyświetlenie wykresu
plt.show()
```
