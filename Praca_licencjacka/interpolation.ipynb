```python
import numpy as np
from scipy.interpolate import CubicSpline
import pandas as pd

# Dane wejściowe
n_values = [
    1, 4, 7, 10, 13, 16, 19, 22, 25, 28, 31, 34, 37, 40, 43, 46, 49, 52, 55, 58,
    61, 64, 67, 70, 73, 76, 79, 82, 85, 88, 91, 94, 97, 100, 103, 106, 109, 112, 
    115, 118, 121, 124, 127, 130, 133, 136, 139, 142, 145, 148, 151, 
]
gdp_values = [ 15606518, 15855037, 16066240, 16432730, 16771381, 16738320, 16738180, 17156119, 17601227, 17632606, 17824071, 18141444, 18152056, 18151998, 18150618, 18922179, 19264939, 19381591, 19573445, 19976827, 19846420, 19844554, 20244900, 20492747, 21089643, 21195070, 21516058, 21974096, 22027880, 22023283, 22719402, 23201380, 23223813, 26477241, 26945391, 27747798, 28132570, 28529436, 28428919, 29617215, 30400960, 30568582, 30928912, 31419689, 31458438, 32332274, 33167334, 34001494, 34586533, 34831634, 35464674]


# Tworzenie spline'u kubicznego
cs = CubicSpline(n_values, gdp_values)

# Obliczanie wartości dla każdego n = 1, 2, 3, ..., 148
n_full = np.arange(1, 149)  # Każdy n od 1 do 148
gdp_interpolated = cs(n_full)  # Oblicz wartości na podstawie spline'u

# Wyświetlenie wyników
print("Interpolowane wartości GDP dla każdego n:")
for n, gdp in zip(n_full, gdp_interpolated):
    print(f"n = {n}, GDP = {gdp:.4f}%")
```

```python
import pandas as pd

# Tworzenie DataFrame z wynikami
results_df = pd.DataFrame({"N": n_full, "GDP (%)": gdp_interpolated})

# Zapis do pliku Excel
results_df.to_excel("interpolated_debt.xlsx", index=False)

print("Wyniki zapisano do pliku 'interpolated_gdp.xlsx'.")
```
