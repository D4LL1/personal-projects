```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = pd.read_excel("C:/Users/wkaco/Desktop/Assets_data_ln.xlsx")
y = pd.read_excel("C:/Users/wkaco/Desktop/macro_data_ln.xlsx")
```

```python
x["id"] = range(len(x))
y["id"] = range(len(y))

df = pd.merge(x, y, on="id")
```

```python
df.drop("id", axis=1, inplace=True)
```

```python
correlation_matrix =  abs(df.corr())
```

```python
plt.figure(figsize=(20, 16))  # Rozmiar wykresu w calach (szeroko, wysoko)

# Twrz map ciepln
sns.heatmap(correlation_matrix, fmt=".2f", cmap="coolwarm", cbar=True)

# Tytu wykresu
plt.title("Macierz korelacji", fontsize=20)

# Wywietl wykres
plt.show()
```
